```python
import pandas as pd
import numpy as np

##### Data preprocessing ####


from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split


##### Classification Algorithms ####

from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier

##### Classification Accuracy ####SVC

from sklearn.metrics import accuracy_score,recall_score,confusion_matrix

##### Save Model ####

import pickle

import warnings
warnings.filterwarnings('ignore')

```

```python
data=pd.read_csv('Banking.csv')
data
```

# Missing Data Treatmaent

```python
# Missing data cheak

data.isna().sum()
```

```python
for col in data.columns:
    if data[col].dtype == 'float64' or data[col].dtype == 'int64' or data[col].dtype == 'float32' or data[col].dtype == 'int32':
        median = data[col].median()
        data[col].fillna(median, inplace=True)
    else:
        data = data.fillna(data.mode().iloc[0])
```

```python
lb_en=LabelEncoder()

New_data['MARITAL_STATUS']=lb_en.fit_transform(New_data['MARITAL_STATUS'])
New_data['GENDER']=lb_en.fit_transform(New_data['GENDER'])
New_data['MEMBER']=lb_en.fit_transform(New_data['MEMBER'])
New_data['DefaulterYN']=lb_en.fit_transform(New_data['DefaulterYN'])
New_data['EDUCATION']=lb_en.fit_transform(New_data['EDUCATION'])

New_data.head()
```

# Identify target variable


```python
### Independent variable ###

x= New_data.drop('DefaulterYN',axis=1)

#### Target varible #####

y=New_data.DefaulterYN


```

```python
x.describe()
```

# Feature Scaling

# Feature scaling is done to reduce dominance of large columns

```python
all_x=list(x.columns)
x[all_x]=x[all_x]/x[all_x].max()

x.describe()
```

# Train and Test Data

```python
x_train ,x_test,y_train,y_test = train_test_split(x,y,test_size =0.3,random_state=1234)
```

```python
len(x_train)
```

```python
len(x_test)
```

# Data preprocessing ends HERE

# --------Classification Algorithms---------

Method 1 - Random Forest Classification

```python
model = RandomForestClassifier()

model.fit(x_train,y_train)

threshold=0.5
pred_y=model.predict_proba(x_test)
pred_y=(pred_y[:,1]>=threshold).astype(int)


print('Confusion Matrix: \n',confusion_matrix(y_test,pred_y))

print('Accuracy: \n',
     round(accuracy_score(y_test,pred_y)*100,2),
      "%"
     )

print('Recall: \n',
     round(recall_score(y_test,pred_y)*100,2),
      "%"
     )
```

# Method 2 - Decision Tree Classifier

```python
model = DecisionTreeClassifier()

model.fit(x_train,y_train)

threshold=0.5
pred_y=model.predict_proba(x_test)
pred_y=(pred_y[:,1]>=threshold).astype(int)


print('Confusion Matrix: \n',confusion_matrix(y_test,pred_y))

print('Accuracy: \n',
     round(accuracy_score(y_test,pred_y)*100,2),
      "%"
     )

print('Recall: \n',
     round(recall_score(y_test,pred_y)*100,2),
      "%"
     )

selected_model =model
```

# Method 3 - KNeighbors Classifier


```python
model = KNeighborsClassifier()

model.fit(x_train,y_train)

threshold=0.5
pred_y=model.predict_proba(x_test)
pred_y=(pred_y[:,1]>=threshold).astype(int)


print('Confusion Matrix: \n',confusion_matrix(y_test,pred_y))

print('Accuracy: \n',
     round(accuracy_score(y_test,pred_y)*100,2),
      "%"
     )

print('Recall: \n',
     round(recall_score(y_test,pred_y)*100,2),
      "%"
     )
```
